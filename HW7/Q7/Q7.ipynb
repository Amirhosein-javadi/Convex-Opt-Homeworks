```julia
using Convex, SCS
```

```julia
y1 = Variable(); # l = exp(y1)
y2 = Variable(); # w = exp(y2)
```

```julia
objective = (pi*exp(y2)+2*exp(y1)) +  2 * exp(y1+y2);
```

```julia
constraints = [
                y1 + y2 >= log(300)              
                y1 >= log(20)
                y1 <= log(30)
                y2 >= log(10)
                y2 <= log(20)
                y1 >= y2
                y1 <= log(2) + y2
                ];
```

```julia
problem = minimize(objective, constraints);
```

```julia
solve!(problem, SCS.Optimizer);
```

```julia
problem.optval
```
